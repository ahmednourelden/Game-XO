```python
import random

"""
Classes names: 
-------------
1-player() 
    player().play(self)         :ask player(person) to select box and then update matrix
    player().pc_play(self)      :ask PC (not smart) to select random box and then update matrix 
    player().smart_pc_play(self):ask smart PC to select box ater take decision and then update matrix 

functions names:
---------------
1-reset() : intial & reset all required variables 
2-check_chance(symbol): find opprotiuntiy to win (used to win or prevent wining)
3-print_matrix() :print matrix or game dashboard  after edit or update
4-mode(mode_no) :choose between 4 different mode to start playing
5-playgame() : calling function to start game
6-win_condition: check if player got winnig game or not
"""

    
class player:
    
    @classmethod
    def __init__(self,play_symbol):
        self.play_symbol = play_symbol
    
    @classmethod
    def play(self):
        while True :
            enter=input(f'Player {str(self.play_symbol)} (Preson) :Please select number of box :')
            try:
                if int(enter) in list(o):
                    m[int(enter)-1]=str(self.play_symbol)
                    o.remove(int(enter))
                    print_matrix()
                    break
                else:
                    print('Please enter box number from 1 to 9')
            except ValueError:
                print('Please enter box number from 1 to 9')
    @classmethod
    def pc_play(self): 
        enter=random.choice(o)
        print (f'Player O (PC) : Entered number of Box= {enter}')
        m[int(enter)-1]=str(self.play_symbol)
        o.remove(int(enter))
        print_matrix()

    @classmethod   
    def smart_pc_play(self): 
        global another_symbol
        if self.play_symbol=='X':       another_symbol='O'
        elif self.play_symbol=='O':     another_symbol='X'
        
        check_chance(str(self.play_symbol))
        if take_decision==True :
            print (f'Player {self.play_symbol} (PC) : Entered number of Box =: {win_box}')
            m[int(win_box)]=str(self.play_symbol)
            o.remove(int(win_box)+1) 
            
        elif take_decision==False:
            check_chance(str(another_symbol))
            if take_decision==True :
                print (f'Player {another_symbol} (PC) : Entered number of Box =: {win_box}')
                m[int(win_box)]=str(self.play_symbol)
                o.remove(int(win_box)+1)
            elif take_decision==False:      
                enter=random.choice(o)
                print (f'Player {self.play_symbol} (PC) : Entered number of Box= {enter}')
                m[int(enter)-1]=str(self.play_symbol)
                o.remove(int(enter))
        print_matrix()

def reset():
    global m , o ,result , who_next,p
    # Represent Place of X or O on matrix
    m=['1','2','3','4','5','6','7','8','9'] 
    #Represent avaiable Options
    o=[1,2,3,4,5,6,7,8,9]
    #result repesent if win condtion has been achivived or not
    result=True
    #who_next reresents who should next player
    who_next=1
    #represent probabilities of box selection to win
    p=[(0,1,2),(0,2,1),(1,2,0),(3,4,5),(4,5,3),(3,5,4),(6,8,7),(6,7,8),
      (7,8,6),(0,3,6),(3,6,0),(0,6,3),(1,7,4),(1,4,7),(4,7,1),(2,5,8),
      (5,8,2),(2,8,5),(0,4,8),(4,8,0),(8,0,4),(2,4,6),(2,6,4),(4,6,2)]

    
def win_condition(symbol):
    #c represent win condition where 3 cells of win position have same value X or O
    c=[(0,1,2),(3,4,5),(6,7,8),(0,3,6),(1,4,7),(2,5,8),(0,4,8),(2,4,6)]
    global result    
    for i in range(8):
        if m[c[i][0]]==m[c[i][1]]==m[c[i][2]]==symbol:
            result=True
            break
    else:
        result=False
            
def check_chance(symbol):
    global take_decision
    take_decision=True
    global win_box
    win_box=0
    for i in range(24):
        if m[p[i][0]]==m[p[i][1]]==symbol and m[p[i][2]] !='X' and m[p[i][2]]!='O' :
            print (f'win_box before leave {win_box}and {p[i]}')
            m[p[i][2]]=symbol 
            win_box=p[i][2]
            break
    else:
        take_decision=False
        print (f'win_box leave {win_box}and take deceision is false after else')
          

def print_matrix():
    print (f'--|--|--\n{m[0]} |{m[1]} |{m[2]} \n--|--|--\n{m[3]} |{m[4]} |{m[5]}' )
    print (f'--|--|--\n{m[6]} |{m[7]} |{m[8]} \n--|--|--')


def mode(mode_no):
    print_matrix()
    global who_next
    who_next=1
    for i in range(9):
        if who_next==1:
            if mode_no=='1' or mode_no=='2' or mode_no=='3':  player('X').play()   
            elif mode_no=='4':                                player('X').smart_pc_play()
            win_condition('X')
            if result:
                print (f'Game Over\nplayer {who_next} win')
                break
            else : who_next=2
        elif who_next==2:
            if mode_no=='1' :     player('O').play()
            elif mode_no=='2':    player('O').pc_play()
            elif mode_no=='3':    player('O').smart_pc_play()
            elif mode_no=='4':    player('O').smart_pc_play()
            win_condition('O')
            if result:
                print (f'Game Over\nplayer {who_next} win')
                break
            else :  who_next=1
    if result==False:
        print ("The Result is Draw ")
        
        
def playgame():
    reset()    
    while True :
        print ('please select on of below modes:\n-------------------------------')
        print ('Mode1 :Two persons will play together , Player 1 is X and Player 2 is O')
        print ('Mode2 :Person and PC (random decision & not smart),Player 1 (Person) is X and Player 2 (PC) is O')
        print ('Mode3 :Person and PC (smart PC),Player 1 (Person) is X and Player 2 (PC) is O')
        print ('Mode4 :Two Smart PCs will play together,Player 1 (PC) is X and Player 2 (PC) is O')
        
        
        mode_number = input ('Please Enter Mode Number : ')
        for i in ['1','2','3','4']:
            if mode_number==i:
                mode(i)
                break
        else:
            print('You entered wrong number , Please enter mode number eg: 1 , 2 , 3 , or 4')
        break
        
```
